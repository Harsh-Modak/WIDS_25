```python
def numpy_spiral_order(matrix):
    result = []

    while matrix.size:
        result.extend(matrix[0])
        matrix = matrix[1:]

        if matrix.size == 0:
            break

        result.extend(matrix[:, -1])
        matrix = matrix[:, :-1]

        if matrix.size == 0:
            break

        result.extend(matrix[-1][::-1])
        matrix = matrix[:-1]

        if matrix.size == 0:
            break

        result.extend(matrix[:, 0][::-1])
        matrix = matrix[:, 1:]

    return result
spiral = numpy_spiral_order(arr.copy())
print("Spiral Order:", spiral)
```

```python
import pandas as pd

df = pd.read_csv("vgsales.csv")
df.head()
```

```python
df['Global_Sales'] = (
    df['NA_Sales'] +
    df['EU_Sales'] +
    df['JP_Sales'] +
    df['Other_Sales']
)

df_sorted = df.sort_values(by='Global_Sales', ascending=False)
df_sorted.head()
```

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum()

plt.figure(figsize=(10,6))
genre_sales.plot(kind='bar')
plt.title("Total Global Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Sales (in millions)")
plt.show()
```

```python
gta = df[df['Name'].str.contains("Grand Theft Auto", case=False, na=False)]

gta_filtered = gta[['Name', 'Platform', 'Year']]
gta_filtered['EU + JP Sales'] = gta['EU_Sales'] + gta['JP_Sales']

gta_filtered
```

```python
sales = {
    'North America': gta['NA_Sales'].sum(),
    'Europe': gta['EU_Sales'].sum(),
    'Japan': gta['JP_Sales'].sum(),
    'Other': gta['Other_Sales'].sum()
}

plt.figure(figsize=(7,7))
plt.pie(sales.values(), labels=sales.keys(), autopct='%1.1f%%')
plt.title("GTA Sales by Region")
plt.show()
```
